```python
import scrapy
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
hr_link = "http://www.hrmorning.com/category/hr-management/page/"
```

```python
data_result = pd.DataFrame(columns=['url','title','tag','time','short_text','long_text'])
```

```python
i = 2
while (i<247):
    print ("index: ", i)
    link = hr_link +str(i)+"/"
    print(link)
    try:
        headers = {'User-Agent':'Mozilla/5.0'}
        page = requests.get(link, headers=headers)
        soup = BeautifulSoup(page.content)
        
        for div in soup.findAll("div", attrs={"class": "post"}):
            url = div.find("a", attrs={"rel": "bookmark"})['href']
            title = div.find("a", attrs={"rel": "bookmark"}).contents
            tag = "HR MANAGEMENT"
            time = div.find("span", attrs={"class": "date published time"}).contents
            
            short_text_tag = div.find("div", attrs={"class": "entry-content"})
            short_text = ""
            if short_text_tag:
                short_text = short_text_tag.contents
            long_text = ""
            data_result.loc[len(data_result)] = [url, title, tag, time, short_text, long_text]

        i += 1
    except Exception as e:
        print(e)
        print ("Code: ", page.status_code)
        time.sleep(5)
        continue
```

```python
data_result.to_csv("hr_morning.csv", index=False)
```

```python
import re
def clean_html(raw_html):
    return re.sub(re.compile('<.*?>'), ' ', raw_html)
def remove_special_character(string):
    string_text = clean_html(string)
    string_text = string_text.replace("['", "")
    string_text = string_text.replace("[", "")
    string_text = string_text.replace("[\"", "")
    string_text = string_text.replace("']", "")
    string_text = string_text.replace("\"]", "")
    string_text = string_text.replace("\n", "")
    
    string_text = re.sub(r'[^\x00-\x7F]+',' ', string_text)
    
    return string_text
```

```python
data_result['title'] = data_result.title.apply(lambda X: remove_special_character(X))
data_result['tag'] = data_result.tag.apply(lambda X: remove_special_character(X))
data_result['short_text'] = data_result.short_text.apply(lambda X: remove_special_character(X))
data_result['long_text'] = data_result.long_text.apply(lambda X: remove_special_character(X))
data_result['time'] = data_result.time.apply(lambda X: remove_special_character(X))
```
