```python
import scrapy
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
hr_entrepreneur_link = "https://www.entrepreneur.com/topic/human-resources"
```

```python
entrepreneur_result = pd.DataFrame(columns=['url', \
                                            'title', \
                                            'description', \
                                            'content', \
                                            'tag', \
                                            'published_data'])
```

```python
i = 1
while (i<60):
    print ("index: ", i)
    link = hr_entrepreneur_link +"/"+str(i)
    print(link)
    try:
        page = requests.get(link)
        soup = BeautifulSoup(page.content)
        for div in soup.findAll("div", attrs={"class": "pl pl-floathero"}):
            title = div.find("a", attrs={"data-ga-action": "headline"}).contents
            description = div.find("div", attrs={"class": "deck"}).contents
            date_time = div.find("time").contents
            tag_tag = div.find("a", attrs={"data-ga-action": "kicker"})
            if tag_tag:
                tag = tag_tag.content
            else:
                tag = ""
            url = div.find("a", attrs={"data-ga-action": "headline"})['href']
            entrepreneur_result.loc[len(entrepreneur_result)] = \
            [url, title, description, "", tag, date_time]

        i += 1
    except Exception as e:
        print(e)
        print ("Code: ", page.status_code)
        time.sleep(5)
        continue
```
